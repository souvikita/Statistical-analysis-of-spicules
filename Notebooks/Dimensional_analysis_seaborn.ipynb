```python
kde_v2 = ds_dict.get_kde_scatter(xax="deform", yax="area_um",kde_type='multivariate',xscale='log',yscale='log')

X, Y, Z = ds_dict.get_kde_contour(xax="deform", yax="area_um",xscale='log',yscale='log',kde_type='multivariate')
Z /= Z.max()

quantiles = [0.13,0.70]
levels = dclab.kde_contours.get_quantile_levels(density=Z,
                                                x=X,
                                                y=Y,
                                                xp=ds_dict["deform"],
                                                yp=ds_dict["area_um"],
                                                q=quantiles,
                                                )
plt.figure(figsize=(8,7))
ax = plt.subplot(111, title="contour lines on KDEs")
sc = ax.scatter(ds_dict["deform"], ds_dict["area_um"], c=kde_v2, marker="o",s=10,cmap='jet')

cn = ax.contour(X, Y, Z,
                levels=levels,
                linestyles=["--", "-"],
                colors=["red", "black"],
                linewidths=[2, 2],
                zorder=2)
ax.set_xscale('log')
ax.set_yscale('log')
plt.colorbar(sc, label="kernel density estimate [a.u]")
fmt = {}
for l, q in zip(levels, quantiles):
    fmt[l] = "{:.0f}th".format(q*100)
plt.clabel(cn, fmt=fmt,fontsize=14)
```
