```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sbsubplot as sfg
import matplotlib.gridspec as gridspec
#import seaborn as sns; sns.set()
```

## RBE analysis 

```python
dpath_npz = '/mn/stornext/d9/souvikb/K_means_results/NUMPY_npz_files/'
data1 = np.load(dpath_npz+'rbe_stats_parallel_improved_shadow_RBEs.npz',allow_pickle=True)
```

```python
new_data1 = data1['arr_0']
ww_list = list(new_data1)
```

```python
max_length = [k[0] for k in ww_list]
eccen = [k[2] for k in ww_list]
max_area =[k[1] for k in ww_list]
```

```python
np.where(np.isnan(max_length)==True)
```

```python
final_eccen=[]
final_area=[]
consider_index =[]
for ii in range(len(eccen)):
    try:
        e = eccen[ii][0]
        final_eccen.append(e)
        a = max_area[ii][0]
        final_area.append(a)
        consider_index.append(ii)
    except:
        print('IndexError at',format(ii))
```

```python
final_max_length=[]
for jj in range(len(consider_index)):
    final_max_length.append(max_length[consider_index[jj]])
```

```python
final_max_length = np.array(final_max_length)
final_eccen = np.array(final_eccen)
final_area = np.array(final_area)
```

## RRE analysis

```python
data2 = np.load(dpath_npz+'rbe_stats_parallel_improved_shadow_RREs.npz',allow_pickle=True)
```

```python
new_data2 = data2['arr_0']
ww_list2 = list(new_data2)
```

```python
max_length_rre = [k[0] for k in ww_list2]
eccen_rre = [k[2] for k in ww_list2]
max_area_rre =[k[1] for k in ww_list2]
```

```python
np.where(np.isnan(max_length_rre)==True)
```

```python
final_eccen_rre=[]
final_area_rre=[]
consider_index_rre =[]
for ii in range(len(eccen_rre)):
    try:
        e = eccen_rre[ii][0]
        final_eccen_rre.append(e)
        a = max_area_rre[ii][0]
        final_area_rre.append(a)
        consider_index_rre.append(ii)
    except:
        print('IndexError at',format(ii))
       # del_index.append(ii)
```

```python
final_max_length_rre=[]
for jj in range(len(consider_index_rre)):
    final_max_length_rre.append(max_length_rre[consider_index_rre[jj]])
```

```python
final_max_length_rre = np.array(final_max_length_rre)
final_eccen_rre = np.array(final_eccen_rre)
final_area_rre = np.array(final_area_rre)
```

### A jointgrid plot between area vs length

```python
#The bottom one is the best one
g = sns.JointGrid(x=final_area_rre, y=final_max_length_rre,xlim=[0.002,final_area_rre.max()],ylim=[0.045,final_max_length_rre.max()],space=0.05)
g.plot_joint(plt.scatter, color='red', edgecolor='red',s=0.5,alpha=0.5)
ax = g.ax_joint
ax.set_xscale('log')
ax.set_yscale('log')
#g = g.plot_marginals(sns.kdeplot, shade=True)
g.plot_marginals(sns.distplot, hist=True, kde=False, color='red',bins=np.logspace(-2.54,np.log10(10),24))
#g.ax_marg_x.hist(final_area_rre, color="b", alpha=.6,
                     #bins=np.logspace(-2.54,np.log10(10),500))
#g.ax_marg_y.hist(final_max_length_rre, color="r", alpha=.6,
                     #orientation="horizontal",
                     #bins=np.logspace(-2.54,np.log10(10),500))
#g = g.plot_marginals(sns.kdeplot, shade=True,color='blue')


g.ax_marg_x.set_xscale('log')
g.ax_marg_y.set_yscale('log')
g.set_axis_labels('Area [Mm$^{2}$]','Length [Mm]')
ax.text(0.05,0.9,'(b)',fontsize=14, transform=ax.transAxes)
```

### A jointgrid plot between length vs. area 

```python
#The bottom one is the best one
g1 = sns.JointGrid(x=final_max_length_rre, y=final_area_rre,xlim=[0.045,final_max_length_rre.max()],ylim=[0.002,final_area_rre.max()],space=0.05)
g1.plot_joint(plt.scatter, color='red', edgecolor='red',s=0.5,alpha=0.5)
ax = g1.ax_joint
ax.set_xscale('log')
ax.set_yscale('log')
#g = g.plot_marginals(sns.kdeplot, shade=True)
g1.plot_marginals(sns.distplot, hist=True, kde=False, color='red',bins=np.logspace(-2.54,np.log10(10),24))
#g.ax_marg_x.hist(final_area_rre, color="b", alpha=.6,
                     #bins=np.logspace(-2.54,np.log10(10),500))
#g.ax_marg_y.hist(final_max_length_rre, color="r", alpha=.6,
                     #orientation="horizontal",
                     #bins=np.logspace(-2.54,np.log10(10),500))
#g = g.plot_marginals(sns.kdeplot, shade=True,color='blue')


g1.ax_marg_x.set_xscale('log')
g1.ax_marg_y.set_yscale('log')
g1.set_axis_labels('Length [Mm]','Area [Mm$^{2}$]')
ax.text(0.05,0.9,'(d)',fontsize=14, transform=ax.transAxes)
```

### A jointgrid between eccentricity vs. length

```python
plt.style.use('seaborn-darkgrid')
fig = plt.figure(figsize=(12,6))
#fig.subplots_adjust(wspace=0.05,left=0.1,right=0.9,top=0.9,bottom=0.1)
gs = gridspec.GridSpec(1, 2, figure=fig)
mg0 = sfg.SeabornFig2Grid(g, fig, gs[0, 0])

f2_ax2 = fig.add_subplot(gs[0, 1])
n1, bins1 = np.histogram(final_max_length_rre,bins= 200)
n1 = n1/np.sum(n1)
bins1 = bins1[:-1]
f2_ax2.plot(bins1, n1,drawstyle='steps-mid',color='red',label='RRE')
f2_ax2.fill_between(bins1,n1,step='mid',alpha=0.4,color='red')
#axs[0].set_xlim(-0.01,2.5)
#f2_ax2.set_ylabel('Probability',fontsize=13)
f2_ax2.spines['right'].set_visible(False)
f2_ax2.spines['top'].set_visible(False)
f2_ax2.axvline(x=0.037*0.722*2,color='black',linestyle='dashed')
#f2_ax2.text(.9, .9, '(a)', transform=gs[1].transAxes, size=14,color='black')
f2_ax2.set_xscale('log')
plt.tight_layout()
```
